```python
import pandas as pd
df = pd.read_csv (r'Downloads/LaptopSales.csv')
mean1 = df['Retail Price'].mean()
sum1 = df['Retail Price'].sum()
max1 = df['Retail Price'].max()
min1 = df['Retail Price'].min()
count1 = df['Retail Price'].count()
median1 = df['Retail Price'].median() 
std1 = df['Retail Price'].std() 
var1 = df['Retail Price'].var()  
print ('Mean price ' + str(mean1))
print ('Sum of price: ' + str(sum1))
print ('Max price: ' + str(max1))
print ('Min price: ' + str(min1))
print ('Count of prices: ' + str(count1))
print ('Median price: ' + str(median1))
print ('Std of prices: ' + str(std1))
print ('Var of sprice: ' + str(var1))
# Source: https://datatofish.com/use-pandas-to-calculate-stats-from-an-imported-csv-file/
groupby_mean1 = df.groupby(['Store Postcode']).mean() 
groupby_count1 = df.groupby(['Store Postcode']).count()
groupby_mean2 = df.groupby(['Integrated Wireless?']).mean() 
groupby_count2 = df.groupby(['Integrated Wireless?']).count()
print ('Sum of values, grouped by store: ' + str(groupby_mean1))
print ('Sum of values, grouped by wireless?: ' + str(groupby_mean2))
```
